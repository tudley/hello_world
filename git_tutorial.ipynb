```python
#   Git tutorial:

#   Commands used:

    git --version  
returns the version of git we're using

    git config --global user.name
    git config -- global user.email
sets the users username and email. Useful for when connecting to github

    git init
initialised a git repository in the current working directory

    git status (optional argument '-- short' give a concise version)
checks the status of the current branch

    git add 'filename' (optional agrument '--all' adds all files in repository)
git adds the file to the staging envinronment 

    git commit -m 'commit message'
git commits the current staged files with a message describing the updates

    git log
provides a log of the commits for a repository

    git help -all
    git <command> -help
provides the options for a command, and prives all commands available

    git branch <branchname>
creates a new branch

    git branch (argument '-b' creates and switches to the branch)
shows all branches

    git checkout <branchname>
switches to a branch

    git merge <branchname>
merges branchname to current branch




#   Installing git

First of, install git from https://www.git-scm.com/
To ensure it's installed correctly, in cmd write:
    
    git --version

and you'll recieve an appropriate ouput, showing git is working correctly.
now we will configure git. This lets git know who you are.
it uses the following information/commands:

    git config --global user.name "<username>"
    git config --global user.email "<email address>"

now for this example, we will create a folder for our project, as I have done already, called git_learning
now navigate to that folder in cmd, and initialize git in the folder through:

    git init   

which responds with:

    Initialized empty Git repository in D:/python_learning/crash_course_exercises/git_learning/.git/

confirming we have initialised an empy Git repository.

Now lets add some files to this git repository.
I'll go ahead an create a python file whihc prints 'Hello world!', called hello.py inside this repository.

now when we call 'dir' in this directory, this notebook and our hello.py files are shown.
we can now check the Git 'status' to see if it's a part of our repo through:

    git status

where we recieve:

    On branch master

    No commits yet

    Untracked files:
    (use "git add <file>..." to include in what will be committed)
            git_tutorial.ipynb
            hello.py

    nothing added to commit but untracked files present (use "git add" to track)

This shows us git is aware of these files, but they have not beed added to our repo.
Files in your git repo can be in one of two states:

    1. Tracked - files that git knows about and are added to the repo
    2. Untraked - files in your working directory, but now added to the repository

When you add files to an empty repository, they all all untracked. To get git to track them, you need to stage them, or add them to the staging environment.


#   Git staging environment

One fo the core functions of the Git is the copcepts of the Staging environment, and the Commit.
As you are working, you may be adding editing and removing files.
But whenever you hit a milestone, should should add the files to the staging environment. 
Staged files are files that are ready to be committed to the repository you're working on. 
for now, we are fone working with hello.oy, so we can add it to the staging environment through:

        git add hello.py

This file should now be stages.
Lets check the status:

    git status

    [OUTPUT]:
    No commits yet

    Changes to be committed:
    (use "git rm --cached <file>..." to unstage)
            new file:   hello.py

Now the file has beed added to the staging environment
Lets add some more files to the working folder. a readme.md, this .ipynb file. Lets also update our .py file to add mroe functionality.
Now, to add all these files to the current directory staging environment:

    git add --all

now when we call git status:

    No commits yet

    Changes to be committed:
    (use "git rm --cached <file>..." to unstage)
            new file:   git_tutorial.ipynb
            new file:   hello.py
            new file:   readme.md

All our files ahve been added to the staging envinronment.
Now we're ready for our first commit.

#   Git commit

Since we have finished out work, we are ready to move from stage to commit for our repo.
Adding comments keep track of our progress and changes are we work.
Git considers each commit change poitn or save point. It is a poit in your project you can go back to if you find a bug, or want to make a change.
when we commit, we should always include a message:

    git commit -m "First commit of hello world"

the command 'commit' performs a commit, and the '-m' adds a messsage

you can also commit without stage. 
Say you make a small change, using the staging env may be a waste of time
we can make a small change to hello.py, when check the status in short format, through:

    git status -short

    [OUTPUT]
     M hello.py
    
Telling us the file is modified.
lets commit directly, through:

    git commit -a -m "updated a new line to hello.py"

#   Git commit log

To view the history of commits, you can use the log command

    git log


# GIT BRANCHES

In git, a branch is a new/serperate version of the main repository.
Lets say you have a large project, and you need to update the design on it.
braches allow you to work on different parts of a project without impacting the main brach.
When the work is complete, a branch can be merged with the main project.
You can even switch between branches and work on different projects without them interfering with eachother.

Lets add some new features to out hello.py
we are working in our local repositorym and we dont want to disturb our main project, so lets create a branch.
to create a branch, call:

    git branch hello.py_user_input

now we can confirm we have created a new branch by calling:

    git branch

    [OUTPUT]

      hello.py_user_input
    * master

we can see we have a master branch, and a new branch we have just created.
The * indicates which branch we are currenty in.
we can use the checkout command to check out a branch, movin us from the current back to the one specified:

    git checkout hello.py_user_input

we get a confirmation we have chaged branches. Now we can make changes to our files.
we can even add another file, goodbye.py

now we check the status of the current branch:

    git status

    [OUTPUT]
    On branch hello.py_user_input
    Changes not staged for commit:
    (use "git add <file>..." to update what will be committed)
    (use "git restore <file>..." to discard changes in working directory)
            modified:   git_tutorial.ipynb
            modified:   hello.py

    Untracked files:
    (use "git add <file>..." to include in what will be committed)
            goodbye.py

Git tells us which branch we're on.
Tells us which files have changes
Tells us which files are untracked.

Lets add everything to the staging environment:

    git add --all

and check the status againL

    git status

everything looks good, so we will commit them to branch.


    git commit -m "added user input to hellp.py, introduces goodbye.py with a function we import into hello.py"


# SWITCHING BETWEEN BRANCHES

Now lets see how we can switch between branches.
looking at the directory contents in the current branch shows:

    8/02/2025  12:41             7,166 git_tutorial.ipynb
    28/02/2025  12:41                48 goodbye.py
    28/02/2025  12:39               231 hello.py
    28/02/2025  11:25               149 readme.md


now if we change to master:

    28/02/2025  12:46             3,940 git_tutorial.ipynb
    28/02/2025  12:46               160 hello.py
    28/02/2025  11:25               149 readme.md

our new files are no longer there. if we open hello.py, we notice our changes are not applied.


#   EMERGENCY BRANCHES

say we are not done with our second branch, but there is a bug in master, i font want to mess with master directly, and i dont want to mess with the branch created, I create a new branch to deal with the EMERGENCY

    git checkout -b emergency-fix

now we have created a new branch from master, and changed to it. Now we can safely fix the error without disturbing the other branches.
Once we have made our changes, we need to get those changes to the master branch.
Check the status:

    git status

as we see the updated files, we can now stage and commit them:

    git add --all
    git commit -m "updated hello.py with emergency fix"

now we have a fix ready for master, we need to merge the two branches.
Lets add some more files to the working folder. a readme.md, this .ipynb file. Lets also update our .py file to add mroe functionality.
Now, to add all these files to the current directory staging environment:

    git add --all

now when we call git status:

    No commits yet

    Changes to be committed:
    (use "git rm --cached <file>..." to unstage)
            new file:   git_tutorial.ipynb
            new file:   hello.py
            new file:   readme.md

All our files ahve been added to the staging envinronment.
Now we're ready for our first commit.

#   Git commit

Since we have finished out work, we are ready to move from stage to commit for our repo.
Adding comments keep track of our progress and changes are we work.
Git considers each commit change poitn or save point. It is a poit in your project you can go back to if you find a bug, or want to make a change.
when we commit, we should always include a message:

    git commit -m "First commit of hello world"

the command 'commit' performs a commit, and the '-m' adds a messsage

you can also commit without stage. 
Say you make a small change, using the staging env may be a waste of time
we can make a small change to hello.py, when check the status in short format, through:

    git status -short

    [OUTPUT]
     M hello.py
    
Telling us the file is modified.
lets commit directly, through:

    git commit -a -m "updated a new line to hello.py"

#   Git commit log

To view the history of commits, you can use the log command

    git log


# GIT BRANCHES

In git, a branch is a new/serperate version of the main repository.
Lets say you have a large project, and you need to update the design on it.
braches allow you to work on different parts of a project without impacting the main brach.
When the work is complete, a branch can be merged with the main project.
You can even switch between branches and work on different projects without them interfering with eachother.

Lets add some new features to out hello.py
we are working in our local repositorym and we dont want to disturb our main project, so lets create a branch.
to create a branch, call:

    git branch hello.py_user_input

now we can confirm we have created a new branch by calling:

    git branch

    [OUTPUT]

      hello.py_user_input
    * master

we can see we have a master branch, and a new branch we have just created.
The * indicates which branch we are currenty in.
we can use the checkout command to check out a branch, movin us from the current back to the one specified:

    git checkout hello.py_user_input

we get a confirmation we have chaged branches. Now we can make changes to our files.
we can even add another file, goodbye.py

now we check the status of the current branch:

    git status

    [OUTPUT]
    On branch hello.py_user_input
    Changes not staged for commit:
    (use "git add <file>..." to update what will be committed)
    (use "git restore <file>..." to discard changes in working directory)
            modified:   git_tutorial.ipynb
            modified:   hello.py

    Untracked files:
    (use "git add <file>..." to include in what will be committed)
            goodbye.py

Git tells us which branch we're on.
Tells us which files have changes
Tells us which files are untracked.

Lets add everything to the staging environment:

    git add --all

and check the status againL

    git status

everything looks good, so we will commit them to branch.


    git commit -m "added user input to hellp.py, introduces goodbye.py with a function we import into hello.py"


# SWITCHING BETWEEN BRANCHES

Now lets see how we can switch between branches.
looking at the directory contents in the current branch shows:

    8/02/2025  12:41             7,166 git_tutorial.ipynb
    28/02/2025  12:41                48 goodbye.py
    28/02/2025  12:39               231 hello.py
    28/02/2025  11:25               149 readme.md


now if we change to master:

    28/02/2025  12:46             3,940 git_tutorial.ipynb
    28/02/2025  12:46               160 hello.py
    28/02/2025  11:25               149 readme.md

our new files are no longer there. if we open hello.py, we notice our changes are not applied.


#   GIT BRANCH merge

We have the emergency fix ready, so lets merge the master and emergency fix branches
First, we need to change to master branch:

    git checkout master

now we merge the current branch (master) witht he emergenxy-fix branch:

    git merge emergency-fix

since the emergency-fix branch came directly from master, and no other changes to master have been made while we were working, git sees this as a continuation from master.
we can now delete emergency-fix branch, as it has been merged:

    git branch -d emergency-fix

Now we can mvoe to the <user_input> branch and continue working there.
lets add some more functionality to hello.py

#   EMERGENCY BRANCHES

say we are not done with our second branch, but there is a bug in master, i font want to mess with master directly, and i dont want to mess with the branch created, I create a new branch to deal with the EMERGENCY

    git checkout -b emergency-fix

now we have created a new branch from master, and changed to it. Now we can safely fix the error without disturbing the other branches.
Once we have made our changes, we need to get those changes to the master branch.
Check the status:

    git status
```
